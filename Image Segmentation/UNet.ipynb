II) Cấu trúc mạng Unet:


![alt text](https://i1.wp.com/nttuan8.com/wp-content/uploads/2019/05/U-Net.png?w=800&ssl=1)

- Tư tưởng của để xây dụng lên mạng Unet đó là nhận thấy CNN có thể phân tách bức ảnh ra thành các feature map mang đặc trưng thông tin của bức ảnh đó vậy thì tại sao lại không sử dụng những đặc trưng nhungnwx thông tin đã được học ấy vào để có thể khôi phục bức ảnh. Với việc sử dụng cùng với những feature maps tương ứng để mở rộng vector thành segmented image. Nó sẽ đảm bảo tính nguyên vẹn bức ảnh tốt hơn ít méo mó đi
Cấu trúc mạng UNet được cấu thành từ 2 phần đó là phần Contracting path và phần Expansive path (có thể hiểu 2 phần này là mã hóa bức ảnh và giải mã bức ảnh)

Với phần Contracting path :
- Theo kiểu của convolutional network với cứ 2 lớp conv2D 3*3 và 1 lớp Maxpooling với nhiệm vụ giảm chiều.
- Đây là phần với nhiệm vụ là nhằm mục đích phân tích bước ảnh tìm ra những thông tin đặc trưng của bức ảnh, học những chi tiết phức tạp hơn.

Sau khi qua phần contracting path lúc này ta biết được bức ảnh nó là gì nhưng lại mất dần đi thông tin về vị trí của nó và để khôi phục lại những vị trí đó ta có phần tiếp theo.

Với phần Expansive path(Phần cốt lõi):
- sau khi đã cô đọng những thông tin vào các feature map ta bắt đầu khôi phục lại kích cỡ bằng cách mở rộng các chiều của phần feature map đã được cô đọng lại trước đó nhằm định vij ra thông tin về vị trí của các pixel.
- Công việc này ngược lại số với phần Contracting path với mỗi block ta dùng 2 conv2D 3*3 kết hợp với Relu sau đó dùng 1 lớp up-convolution 2*2( UpSampling) kết hợp với conv2D 2*2 để tăng dần kích thước ảnh và giảm chiều sâu của ảnh.
- như đã nói từ trước là những feature map bên phía Contracting path mang những thông tin, đặc tính của ảnh nên ta sẽ sử dụng những feature maps tương ứng với những lần upsampling để có thể khôi phục lại ảnh tốt hơn nó bảo tồn được nhiều đặc tính của ảnh hơn.

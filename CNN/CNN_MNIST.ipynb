```
from keras.preprocessing.image import ImageDataGenerator
import glob
import numpy as np
import cv2 as cv
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from keras.utils import to_categorical
import scipy.io
np.random.seed(2)
mat = scipy.io.loadmat('/content/drive/My Drive/train_32x32.mat')
x_train = mat['X']
y_train = mat['y']
print(x_train.shape)
print(y_train.shape)
x = x_train/255.0

print(x.shape)
# x = np.array(x).reshape(73257,32,32,3)
train = []
for i in range(73257):
  img = x[:,:,:,i]

  train.append(img)
# print(x[:,:,:,1].shape)
def rgb2gray(images):
    return np.expand_dims(np.dot(images, [0.2990, 0.5870, 0.1140]), axis=3)

preX = []
preY = []
gray = rgb2gray(train)

print(gray.shape)
for i in range(73257):
  x = y_train[i]
  # print(x)
  if  x == 10:
    y_train[i] = 0
train = np.array(train).reshape(-1,32,32,1)
print(gray.shape)
y = to_categorical(y_train)
# print(y_train[i])
# plt.imshow(gray[1], cmap='gray')
print(y.shape)
for i in range(73257):
  preX.append(gray[i])
  preY.append(y[i])
# plt.show()
for i in range(20000):
  preX.append(x_mnist[i])
  preY.append(y_mnist[i])
preX = np.array(preX)
print(preX.shape[0])
per = np.random.permutation(preX.shape[0])

dataX =[]
dataY = []

for i in per:
  dataX.append(preX[i])
  dataY.append(preY[i])
dataX=  np.array(dataX)
dataY = np.array(dataY)
X_train, X_val, Y_train, Y_val = train_test_split(dataX, dataY, test_size = 0.1,random_state = 2)
print(X_val.shape,Y_val.shape)
data_gen = ImageDataGenerator(zoom_range = 0.2,
                              width_shift_range = 0.05,height_shift_range = 0.05,
                              rotation_range = 20)

data_gen.fit(X_train)
```
